#**Visualização de Dados com Matplotlib e Seaborn**

#Importando Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#Importando Banco de Dados

```python
url = 'https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv'
```

```python
df = pd.read_csv(url)
df
```

```python
df.info()
```

#Pré-processamento

```python
#colocando o país como índice
df.set_index('País', inplace=True)
```

```python
#organizando a variável ano
anos = list(map(str, range(1980, 2014)))
anos
```

```python
#filtrando as informações do Brasil
brasil = df.loc['Brasil', anos]

#criando um dicionário para o dataframe
brasil_dict = {'ano':brasil.index.tolist(), 'imigrantes':brasil.values.tolist()}

#organizando as informações em um dataframe
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

#Visualização dos Dados

##Tendências de imigração do Brasil para o Canadá, no período de 1980 a 2013

```python
plt.figure(figsize=(8,4)) #formatando o tamanho do gráfico
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes']) #selecionando os dados
plt.title('Imigração do Brasil para o Canadá') #colocando um título
plt.xlabel('Ano') #legendando eixo x
plt.ylabel('Número de Imigrantes') #legendando eixo y
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) #formatando o eixo x para mostrar de 5 em 5 anos
plt.show() #não mostra mensagem ao executar o código
```

```python
fig, ax = plt.subplots(figsize=(8,4)) #formatando o tamanho do gráfico
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g') #selecionando os dados
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #formatando o eixo x para mostrar de 5 em 5 anos
plt.grid(linestyle='--')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#salvando gráfico
fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

Gráfico de linha x Boxplot da Imigração do Brasil para o Canadá

```python
fig, axs = plt.subplots(1,2, figsize = (15,5))

axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()
```

##Comparação da Imigração dos 4 maiores países da América do Sul: Brasil, Colômbia, Argentina e Peru

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3) #ajustando os gráficos para não ter sobreposição
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\n1980 a 2013') #adicionando um tútulo geral

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#formatando o eixo x de todos os gráficos para mostrar de 5 em 5 anos
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

#legendando eixo x e eixo y de todos os gráficos
for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

#atribuindo o valor máximo e mínimo ao eixo y
ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

##Comparação Brasil x Países América do Sul

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
#lista com as cores que serão utilizadas
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
#destacando o Brasil com a cor verde e os demais com cinza
cores = []

for pais in america_sul_sorted.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

#gráfico de barras horizontal
fig, ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_ylabel('')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

#adicionando rótulos de valores nas barras
for i, v in enumerate(america_sul_sorted['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False) #retirando a caixa em volta do gráfico
ax.get_xaxis().set_visible(False) #retirando eixo x
ax.tick_params(axis='both', which='both', length=0) #retirando os ticks

#salvando gráfico
fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

Testando paletas de cores no gráfico

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', hue='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')
    ax.legend().remove() #tirando o campo de legenda
    plt.show()
```

```python
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```

```python
gerar_grafico_paleta('coolwarm') #mais indicado para dados que girem em torno do valor central
```

```python
gerar_grafico_paleta('tab10') #indicado para valores categóricos
```

Alterando o tema do gráfico

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

Retirando as bordas do gráfico

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='tab10')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
sns.despine()

plt.show()
```

##Criando gráficos interativos

```python
fig = px.line(dados_brasil, x='ano',y='imigrantes',
              title='Imigração do Brasil para o Canadá no período de 1980 a 2013')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(width=1000, height=500,
                  xaxis={'tickangle':-45},
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')
fig.show()
```
